```python
def convert_to_lowercase(x):
    return [i.lower() for i in x]
```

```python
features = ['director_name', 'actor_2_name', 'genres', 'actor_1_name', 'actor_3_name', 'plot_keywords', 'language']
for feature in features:
    data[feature] = convert_to_lowercase(data[feature])
data.head()
```

```python
def remove_hyphens(x):
    return [i.replace("-", "") for i in x]

def remove_bars(x):
    return [i.replace("|", " ") for i in x]
```

```python
columns = ['genres', 'plot_keywords']

for column in columns:
    data[column] = remove_hyphens(data[column])
    data[column] = remove_bars(data[column])  
```

```python
def remove_xa0(x):
    return [i.replace("\xa0", "") for i in x]
```

```python
data['movie_title'] = remove_xa0(data['movie_title'])
```

```python
data.head()
```

```python
def create_feature(x):
    return x['plot_keywords'] + ' ' + x['actor_1_name'] + ' ' + x['actor_2_name'] + ' ' + x['actor_3_name'] + ' ' + x['director_name'] + ' ' + ' ' + x['genres'] + ' ' + x['language'] #+ ' ' + x['movie_title']
```

```python
data['final_feature'] = data.apply(create_feature, axis=1)
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(data['final_feature'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
indices = pd.Series(data.index, index = data['movie_title']).drop_duplicates()
```

```python
def get_recommendations(title, cosine_sim = cosine_sim):
    idx = indices[str(title)]
    print(idx)

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:11]

    movie_indices = [i[0] for i in sim_scores]

    return data['movie_title'].iloc[movie_indices]
```

```python
get_recommendations('The Dark Knight Rises')
```

```python
get_recommendations('The Hateful Eight')
```

```python
get_recommendations('Spectre')
```
